```python
from __future__ import division, print_function, unicode_literals
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
a = pd.read_csv('Salary_dataset.csv', sep=',')
X = a['YearsExperience']
y = a['Salary']

plt.plot(X, y, 'ro')
plt.xlabel('Year')
plt.ylabel('salary')
plt.show()
```

```python
# Building Xbar
one = np.ones((X.shape[0]))
Xbar = np.concatenate((one, X), axis=0).reshape(-1,2,order='F')


# Calculating weights of the fitting line
A = np.dot(Xbar.T, Xbar)
b = np.dot(Xbar.T, y)
w = np.dot(np.linalg.pinv(A), b)
print('w = ', w)
# Preparing the fitting line

```

```python
w_0 = w[0]
w_1 = w[1]
x0 = np.linspace(0, 12, 2)
y0 = w_0 + w_1 * x0

# Drawing the fitting line
plt.plot(X.T, y.T, 'ro')  # data
plt.plot(x0, y0)  # the fitting line
plt.xlabel('Year')
plt.ylabel('Salary')
plt.show()
```
